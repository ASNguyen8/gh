```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def distance(a,b):
    return ( (a[0]-b[0])**2+(a[1]-b[1])**2 )**.5
```

```python
x=(0,0) #L'origine du plan
y=(1,1) #Un certain point du plan

print("Le point x de coordonnées",x,"et le point y de coordonnées",y,"sont séparés de la distance ",distance(x,y))
```

```python
assert 2**.5==distance(x,y)
```
